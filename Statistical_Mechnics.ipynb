### Problem 1

```python
N=[100,500,1000]

import random
import numpy as np
import matplotlib.pyplot as plt 
t=0
S=0
S1=0
n_list=[]
m_n=[]
m_n2=[]
sig=[]
for i in range(len(N)):
  n=N[i]
  while t<10000:
    r = random.uniform(0,1)
    if r<= n/N[i]:
       n=n-1
    else:
      n=n+1 
    t=t+1
    n_list.append(n)
    S=S+n
    S1=S1+n**2
    m_n.append(S/10000)
    m_n2.append(S1/10000)
    sig.append((m_n2[i]-m_n[i]**2)**0.5)



T=np.arange(1,10000+1,1)
n_list=np.asarray(n_list)
for i in range(len(N)):
  plt.plot(T,n_list/N[i])
  plt.show()
  
 

plt.title('n vs t')
plt.xlabel('t')
plt.ylabel(r'$\frac{n}{N}$')
plt.show()
for i in range(len(N)):
  print("Number of particles : ",N[i], ' | <n> : ',m_n[i],' | sigma/N : ',sig[i]/N[i])
```

```python
n=[i for i in range(4)]
c=[0 for i in range(4)]
nmax=[100,500,1000,5000]
obs=0
for j in range(len(nmax)):
  while obs<nmax[j]:
    n1=0
    for i in range(3):
      r = random.uniform(0,1)
      if r>=0.5:
        n1+=1
    c[n1]=c[n1]+1
    obs+=1 
  print("No. of Observations : ",nmax[j])  
  sum=0
  for i in range(len(c)):
    sum+=c[i]
  for i in range(len(c)):
    print("probabily of getting %d heads : "%n[i],c[i]/sum)
```
